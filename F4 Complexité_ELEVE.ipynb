# <span style="color: #337dbd">Notion de complexité d'un programme</span>

La complexité d’un problème mathématique P est une mesure de la quantité de
ressources necessaires à la résolution du problème P.

Cette mesure est basée sur une estimation du nombre d’opérations de base
effectuées par l’algorithme en fonction de la taille **n** des données en entrée de
l’algorithme.

Généralement, pour le même problème P, on peut trouver plusieurs algorithmes
(Alg 1 , Alg 2 , .., Alg n ).

L’objectif de la complexité est d’évaluer le **coût** d’exécution de chaque
algorithme afin de choisir le meilleur.

<BLOCKQUOTE style='background-color:#abd3d0;'>    

**NOTE :**
Chaque opération élémentaire a un coût qui vaut 1 :
    
1) Affectation 
    
2) Test de comparaison : ==, <, <=, >=, ! = 
    
3) Opération de lecture (input) et écriture (print) 
    
4) Opération arithmétique : +, −, ∗, /, % 
    
5) Opération d’incrémentation : a=a+1 (pas de 1), a=a+n (pas de n) 
    
6) Opération de décrémentation : a=a-1 (pas de 1), a=a-n (pas de n)
    
    
</BLOCKQUOTE>   

Exemple pour le programme 1, si la chaine de caractère passée en argument est de taille 2 avec la lettre "a" en seconde position : 

````python
def contient_a_1(chaine) :            #        COUT           2eme tour
    k=0                               #         1    
    N = len(chaine)                   #         1+1
    result = False                    #         1                  
    while result == False and k<N:    #         1+1              1+1
        if chaine[k] == "a":          #         1                1
            result = True                 
        k=k+1                         #         1                1    
    return result                                          
    
                                      # SOIT UN COUT TOTAL DE 12
````

Exemple pour le programme 2, si la chaine de caractère passée en argument est de taille 2 avec la lettre "a" en seconde position : 

````python
def contient_a_2(chaine) :            #        COUT                    2eme tour
       for i in range(len(chaine)) :  #         1+1               1+1
            if chaine[i] == "a" :     #         1                 1  
                return True
            return False
                                       # SOIT UN COUT TOTAL DE 6 !!
````

Vous avez également remarqué que le temps d'éxécution va dépendre du fait que la lettre "a" sera présente ou non ou suivant qu'elle sera au début ou à la fin de la chaine de carcatères.

<BLOCKQUOTE style='background-color:#abd3d0;'>    

**NOTE :** Ainsi on parlera de : 
     
- de **coût dans le meilleur des cas** : le programme s'éxécute en un minimum d'opérations,
    
- de **coût dans le pire des cas**.: le programme s'éxécute en un maximum d'opérations (ici, parcourir tout le mot),
- **de coût moyen**. 
</BLOCKQUOTE>   

Le temps déxécution d'un programme est corrélé avec son coût c'est pourquoi le programme 1 était moins bon que le 2.

Pour évaluer le caractère efficace ou non d'un programme, on parle de **complexité** d'un programme.

<BLOCKQUOTE style='background-color:#ff870e;'>    

**DEFINITION :**
    
La **complexité** d'un programme, notée **O(f(n))**, est un ordre de grandeur décrit par la fonction **f(n)** avec **n** le nombre de données à traiter quand **n** tend vers + l'infini.
</BLOCKQUOTE>   

### <span style="color: #ce0000">Exemple : </span>  

Si on observe le tracé de la reprsentation graphique du temps d'éxécution de la fonction ```contient_a_1()``` en fonction de a taille **n** de la chaine de caracatère prise en argument :

````python
import time
from matplotlib import pyplot as plt  

def contient_a_1(chaine) :                      # premier programme
    k=0
    N = len(chaine)
    result = False
    while result == False and k<N:
        if chaine[k] == "a":
            result = True
        k=k+1
    return result


temps=[]
taille=[]
mesure =0
for i in range(0,10000,100):                    # taille de la chaine de caractères
    chaine="r"*i                                # chaine de caractères à étudier
    for t in range(100):                        # nombre d'essais pour calculer un temps moyen
        start= time.clock()                     # démarrage de l'horloge
        contient_a_1(chaine)                    # exécution de la fonction
        end= time.clock()                       # arrêt de l'horloge
        mesure= end-start                       # mesure du temps d'éxécution de la fonction
        mesure+=mesure
    moy = mesure/100                            # calcul du temps moyen d'éxécution de 100 fois avec une même chaine
    temps.append(moy)
    taille.append(i)
    
  
plt.plot(taille,temps)                           
plt.grid(True)  
````

Vous reconnaissez j'espère une fonction affine. Cette fonction est dite en complexité **O(n)**.

La complexité est ainsi propostionnelle à la taille de **n**.

Vous devinez sans doute qu'il est préférable d'écrire un programme avec une complexité la plus petite possible.

Le tableau ci-dessous donne un ordre de grandeur des temps d'éxécutions de programmes, suivant le nombre **n** d'opérations et suivant la complexité : en O(log n), O(2n), O(n log n)...
